# Итератор

```python
arr = [1, 2, 8, 6, 44, 23, 97, 3, 2, 7]
arr_iter = iter(arr)
print(arr_iter)
print(next(arr_iter))
print(arr_iter.__next__())
```

```python
r = range(4, 100, 13)
r_iter = iter(r)
print(r_iter)
print(next(r_iter))
print(r_iter.__next__())
```

# Генератор

```python
arr = [x for x in range(1,100) if x % 10 == 0 or x % 12 == 0]
print(arr)
```

```python
def is_simple(n):
    if n == 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

arr = [x for x in range(1,100) if is_simple(x)]
print(arr)
```
