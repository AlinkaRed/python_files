```python
import shutil
import codecs
import sys
l = codecs.open( "t.txt", "r", "utf_8_sig" )
s = 0
c = 0
for i in l:
    q = i.split()
    fname = q[0]
    sname = q[1]
    mark = int(q[2])
    s += mark
    c += 1
    if mark < 3:
        print(fname, sname, mark)
l.close()
print('Средний балл группы:', s/c)
```

```python
class Rectangle:
    def __init__(self, a, b):
        self.a = a
        self.b = b
    def aa(self):
        return self.a
    def bb(self):
        return self.b
    def p(self):
        return (self.b + self.a)*2
    def s(self):
        return self.b * self.a
r = Rectangle(9, 16)
r.s()
```

```python
class Math:
    def __init__(self, a = 0, b = 0):
        self.a = a
        self.b = b
    def addition(self):
        return self.a+self.b
    def substraction(self):
        return self.a-self.b
    def multiplication(self):
        return self.a*self.b
    def division(self):
        return self.a/self.b
a = Math(5, 10)
a.addition()
```

```python
class Soda:
    def __init__(self, a = ''):
        self.a = a
    def add(self):
        if self.a!='':
            return f'Газировка и {self.a}'
        else:
            return 'Обычная газировка'
s = Soda('a')
s.add()
```
