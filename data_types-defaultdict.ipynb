# Defaultdict

```python
from collections import defaultdict

my_dict = defaultdict(list)

my_dict['fruits'].append('apple')
my_dict['vegetables'].append('carrot')

print(my_dict)
```

```python
letters = {}

for letter in "Mississippi":
    if letter not in letters:
        letters[letter] = 1
    else:
        letters[letter] +=1

print(letters)
```

```python
from collections import defaultdict

letters = defaultdict(int)

for letter in "Mississippi":
    letters[letter] += 1

print(letters)
```

```python
dict(letters)
```

```python
from collections import defaultdict

my_word = "Mississippi"

d1 = defaultdict(list)

for index, letter in enumerate(my_word):
    d1[letter].append(index)

print(d1)
```
