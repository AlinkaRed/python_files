```python
a = 0
k = []
for i in s:
    q = '' + i
    m = len(q)
    e = q[::-1]
    if q[::]==e[::] and q.isalpha() and not q in k and m>3:
        a+=1
        k.append(q)
        print(a, ' ', q)
```
