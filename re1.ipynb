```python
import re
import uuid
import random
```

```python
with open('Винокуров Железные рыцари-6.txt', encoding='utf-8') as f:
    s=f.read()
```

```python
for i in range(10):
    uid = uuid.uuid4()
    l = len(s)
    r = random.randint(0, l)
    s = s[:r]+' '+str(uid)+' '+s[r:]
```

```python
file = open("f_uid.txt", mode="w", encoding='utf-8')
file.write(s)
file.close()
```
