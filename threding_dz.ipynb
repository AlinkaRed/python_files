```python
import shutil
import os
import datetime
from concurrent.futures import ThreadPoolExecutor
import time
from time import sleep
from threading import Thread, Lock
from multiprocessing import Process, Pool, Queue

def timer(func):
    def inner(*args, **kwargs):
        start = time.time()
        res = func(*args, **kwargs)
        elapsed = time.time() - start
        print(f'Time: {elapsed:.5f} sec ({func.__name__})')
        return res
    return inner
```

```python
@timer
def sync():
    folder_path = "C:/Users/USER/Desktop/threding_test"
    
    if os.path.exists(folder_path):
        shutil.rmtree(folder_path)
    
    os.makedirs(folder_path)
    
    file_source = "C:/Users/USER/Desktop/threding_test/test.txt"
    
    with open(file_source, "w") as f:
        f.write("1" * 1024 * 1024)
    
    for i in range(10):
        file_copy = f"C:/Users/USER/Desktop/threding_test/test-{i+1}.txt"
        shutil.copy(file_source, file_copy)

sync()
```
