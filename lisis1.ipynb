```python
import time

def timer(fun):
    
    def inner(*args, **kwargs):
        start = time.time()
        res = fun(*args, **kwargs)
        print(time.time() - start)
        return res
    return inner

@timer
def fun(x):
    y = 0
    for i in range(x):
        y += i
    return y
```

```python
class Group:
    def __init__(self, name, num):
        self.name = name
        self.num = num
class Student:
    students = set() 

    def __init__(self, first_name, last_name, group):
        self.first_name = first_name
        self.last_name = last_name
        self.group = group
        if (first_name, last_name) in Student.students:
            raise ValueError('       ')
        Student.existing_students.add((first_name, last_name))


class Mark:
    def __init__(self, student, subject, mark):
        self.student = student
        self.subject = subject
        if mark < 2 or mark > 5:
            raise ValueError(' ')
        self.mark = mark


class Journal:
    def __init__(self):
        self.marks = []

    def add_mark(self, mark):
        self.marks.append(mark)

    def get_group_performance(self, group_num, subject):
        filtered_marks = [
            mark for mark in self.marks
            if mark.student.group.num == group_num and mark.subject == subject
        ]
        filtered_marks.sort(key=lambda x: (x.student.group.num, x.student.last_name, x.subject))
        return filtered_marks

    def get_student_performance(self, first_name, last_name):
        filtered_marks = [
            mark for mark in self.marks
            if mark.student.first_name == first_name and mark.student.last_name == last_name
        ]
        filtered_marks.sort(key=lambda x: x.subject)  #    
        return filtered_marks

    def average_mark_for_group(self, group_num):
        filtered_marks = [
            mark.mark for mark in self.marks
            if mark.student.group.num == group_num
        ]
        if not filtered_marks:
            return 0  #      
        return sum(filtered_marks) / len(filtered_marks)  #  
```

```python
#  
group1 = Group(" ", 1)
group2 = Group(" ", 2)

#  
student1 = Student("", "", group1)
student2 = Student("", "", group1)
student3 = Student("", "", group2)

#  
journal = Journal()

#  
journal.add_mark(Mark(student1, "", 4))
journal.add_mark(Mark(student1, "", 5))
journal.add_mark(Mark(student2, "", 3))
journal.add_mark(Mark(student3, "", 4))

#    1   ""
group_performance = journal.get_group_performance(1, "")
for mark in group_performance:
    print(f"{mark.student.first_name} {mark.student.last_name}: {mark.mark}")

#    " "
student_performance = journal.get_student_performance("", "")
for mark in student_performance:
    print(f"{mark.subject}: {mark.mark}")

#     1
average = journal.average_mark_for_group(1)
print(f"    1: {average}")
```
