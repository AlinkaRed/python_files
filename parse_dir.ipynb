```python
import os
import pandas as pd

def parse_files_to_dataframe(folder_path):
    file_info = []
    for dirpath, dirnames, filenames in os.walk(folder_path):
        for filename in filenames:
            full_path = os.path.join(dirpath, filename)
            file_size = os.path.getsize(full_path) 
            file_info.append({'filename': filename, 'path': full_path, 'size': file_size})
    df = pd.DataFrame(file_info)
    return df

def main(folder_path, res_path = 'res1.xlsx'):
    df_files = parse_files_to_dataframe(folder_path)
    df_files.to_excel('res1.xlsx', index=False)
    print("Данные успешно сохранены в файл")
    print(df_files)
```

```python
import os
import json
import sys
from collections import defaultdict

def parse_filenames_to_json(folder_path, output_file):    
    data = defaultdict(lambda: {
        'id': None,
        'paths': [],
        'name': None,
        'weights': []
    })

    if os.path.exists(output_file):
        with open(output_file, 'r', encoding='utf-8') as f:
            existing_data = json.load(f)
        existing_data = {i['id']: i for i in existing_data}
        data.update(existing_data)
            
    for dirpath, _, filenames in os.walk(folder_path):
        for filename in filenames:
            if filename.endswith('.png'):
                parts = filename[:-4].split('_')
                if len(parts) < 4:
                    continue

                id1 = parts[0]
                n1 = int(parts[1])
                n2 = int(parts[2])
                nameId = str(parts[3])

                data[id1]['id'] = id1
                data[id1]['paths'].append(os.path.join(dirpath, filename))
                data[id1]['name'] = nameId
                data[id1]['weights'].append([n1, n2])

    result = list(data.values())

    with open(output_file, 'w', encoding='utf-8') as f:
        json.dump(result, f, ensure_ascii=False, indent=4)

def main(folder_path, output_file):
    parse_filenames_to_json(folder_path, output_file)
    print(f"Данные успешно сохранены в файл: {output_file}")
    with open(output_file, 'r', encoding='utf-8') as f:
        saved_data = json.load(f)
        print("Содержимое файла JSON:")
        print(json.dumps(saved_data, ensure_ascii=False, indent=4))

if __name__ == "__main__":
    # if len(sys.argv) != 3:
    #      sys.exit(1)

    # folder_path = sys.argv[1]
    # output_file = sys.argv[2]

    main('C:/Users/USER/Desktop/PNG', 'r3.json')
```

Каталог может содержать подкаталоги. В подкаталогах располагаются .pdf файлы.
Модуль принимает два параметра:
путь к каталогу с исходными данными
путь к базовому каталогу для сохранения результатов
Структуру подкаталогов необходимо сохранить!
PDF конвертировать в набор PNG файлов
Имя файла формировать согласно следующему правилу:
базовое имя name.pdf конвертировать в
name_номерлиста.png
Выровнять данные по наиболее протяженной стороне плана помещения.

```python
import os
import pymupdf
import sys

def main(folder_path, output_path):
    for dirpath, _, filenames in os.walk(folder_path):
        for filename in filenames:
            if filename.endswith('.pdf'):
                full_path = os.path.join(dirpath, filename)
                relative_path = os.path.relpath(full_path, folder_path)
                relative_path = relative_path.replace('\\', '/')
                doc = pymupdf.open(full_path)
                
                for page in doc:
                    if page.rect[2] < page.rect[3]:
                        page.set_rotation(270)
                    pix = page.get_pixmap()
                    
                    output_dir = os.path.join(output_path, os.path.dirname(relative_path))
                    os.makedirs(output_dir, exist_ok=True)
                    
                    pix.save(os.path.join(output_dir, f'{os.path.splitext(filename)[0]}_page_{page.number + 1}.png'))

main('C:/Users/USER/Desktop/Пример', 'C:/Users/USER/Desktop/Результ')
```
