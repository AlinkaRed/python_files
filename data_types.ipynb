# Стеки и Очереди

### 1. Стеки - это списки

```python
stack = list('abcde')
stack
```

```python
# методы списков
[i for i in dir(stack) if not i.startswith('_')]
```

```python
# 1. добавление в список
stack.append(0)
print(1, stack)

# 2. расширение списка
stack.extend(list('xyz'))
print(2, stack)
stack += list('123')
print(2.1, stack)

# 3. взять последний элемент (удаляется)
last = stack.pop()
print(3, last, stack)

# 4. взять i-тый элемент (удаляется)
elem = stack.pop(-2)
print(4, elem, stack)

# 5. вставить элемент по индексу
stack.insert(2, 'y')
print(5, stack)

# 6. посчитать к-во определенных элементов
count1 = stack.count('y')
print(6, count1)

# 7. посчитать к-во всех элементов
count2 = len(stack)
print(7, count2)
```

### 2. Очереди (Queue) 

```python
from collections import deque

queue = deque('abcde', maxlen=6)
queue
```

```python
# методы queue
[i for i in dir(queue) if not i.startswith('_')]
```

```python
from collections import deque

queue = deque('abcde')
queue
```

```python
queue.append(0)
print(queue)
```

```python
queue = deque('abcde')
queue.appendleft(0)
print(queue)
```

```python
queue = deque('abcde')
queue.extend(list('xyz'))
print(queue)
queue += list('123')
print(queue)
```

```python
queue = deque('abcde')
queue.extendleft(list('xyz'))
print(2, queue)
queue += list('123')
print(queue)
```

```python
queue = deque('abcde')
last = queue.pop()
print(last, queue)
```

```python
queue = deque('abcde')
last = queue.popleft()
print(last, queue)
```

```python
queue = deque('abcde')
queue.insert(2, 'y')
print(queue)
```

```python
queue = deque('abcdeyy')
count1 = queue.count('y')
print(count1)
```

```python
count2 = len(queue)
print(count2)
```
