```python
import csv
import json
import zipfile, os
```

```python
with open('Винокуров Железные рыцари-6.txt', encoding='utf-8') as f:
    text = f.read()
text = text.strip().split()
a = dict()
```

```python
for t in text:
    t = t.strip(':)?!(,.;…—><"«!»')
    if t.isalpha() and len(t)>4:
        if t in a:
            a[t]+=1
        else:
            a.update({t:1})
```

```python
j = json.dumps(a, ensure_ascii=False)
```

```python
with open('csv_test.csv', 'w') as f:
    f.write(j)
```

```python
path = '.'
file_dir = os.listdir(path)
with zipfile.ZipFile('test.zip', mode='w', compression=zipfile.ZIP_DEFLATED) as zf:
    for file in file_dir:
        if file.endswith('csv_test.csv'):
            add_file = os.path.join(path, file)
            zf.write(add_file)
```
