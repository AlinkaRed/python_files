# Map

```python
s = ['1', '2', '3', '4']
res = map(int, s)
res
```

```python
list(res)
```

```python
s = '125243162739'
res = list(map(int, s))
res
```

```python
def double(val):
    return val * 2

a = [1, 2, 3, 4]    
res = list(map(double, a))
print(res)
```

```python
def double(val):
    return val * 2

a = 'abc'    
res = list(map(double, a))
print(res)
```

```python
a = [1, 2, 3, 4]    
res = list(map(lambda x : x * 2, a))
res
```

```python
a = [1, 2, 3]
b = [4, 5, 6]
res = map(lambda x, y: x + y, a, b)
print(list(res))
```

```python
a = [1, 2, 3]
b = [4, 5, 6]
c = [7, 8, 9]
res = map(lambda x, y, z: x * y * z, a, b, c)
print(list(res))
```

```python
a = ['alina', 'alex']
res = list(map(str.capitalize, a))
res
```

```python
a = ['alina', 'alex']
res = list(map(lambda x : x[0], a))
res
```

# Zip

```python
names = ['John', 'Alice', 'Bob', 'Lucy']
scores = [85, 90, 78, 92]

res = zip(names, scores)
res
```

```python
list(res)
```

```python
a = [1,9,8,3,5]
r = list(zip(a))
r
```

```python
b = 'abcde'
r = list(zip(a,b))
r
```

```python
c = (1,4,5,6,7)
r = list(zip(a,b,c))
r
```

```python
a = [('Apple', 10), ('Banana', 20), ('Orange', 30)]

fruits, quantities = zip(*a)

print(f"Fruits: {fruits}")
print(f"Quantities: {quantities}")
```

```python
d = {'name': 'Alice', 'age': 25, 'grade': 'A'}

keys = d.keys()
values = d.values()

res = zip(keys, values)
print(list(res))
```
