```python
s='''
hgajs jshsja (FISH) . ABBA khshydydhmsn (CAR) YVYUYYFFFi yyuHJgJgsj
PPOJihshdkdld. KITTY loejiwhuhef [DOG] ksisidodkk, RADAR {CAT}
'''
blacklist=['FISH','CAR']
whitelist=['CAT', 'KITTY']
s=s.split()
d=[]
for i in s:
    i=i.strip('.,)({}[]  ')
    w=i.upper()
    p=i[::-1]
    if i==w and i.isalpha() and i not in blacklist and i in whitelist or p==i and i.isalpha():
        d.append(i)
q=', '.join(d)
print(f'words: {q}')
```

```python
text = """
Название.\n
\n
Глава 1. Сказ о бла-бла-бла.\n
\n
   Было это  в  тридевятом царстве,  в тридесятом  государстве, \n
где бы ничего, но  случился  бла-бла-бла.  Решили его пофиксить, \n
но не тут-то было! \n
Ведь бла-бла-бла, да бла-бла-бла, неразделимы, как ни печально...\n
\n
   Бились они, бились,  но, как ни крути, бла-бла-бла оставался \n
бла-бла-бла.\n
\n
\n
Глава 2. Интрига и бла-бла-бла.\n
\n
   Долго  сказка  сказывалась,  но  не долго  дело делалось, ибо\n
бла-бла-бла делать что-то долго.\n
\n
\n
Глава 3. Финал.\n
\n
   И жили они долго и счастливо...\n
\n
   Особенно, бла-бла-бла.  Ведь  бла-бла-бла  как-то сам  собой  \n
прошёл.\n
"""
p = text.count('бла-бла-бла')
for i in range(p):
    if i%2==1:
        text = text.replace('бла-бла-бла','КРИНЖ',1)
    else:
        text = text.replace('бла-бла-бла','ПИНЖ',1)
text = text.splitlines()
n = len(text)
s = ''
a = 0
for i in range(n):
    if text[i]!='':
        if text[i].endswith('.') or text[i].endswith('!'):
            q=' '.join(text[i].split())
            if q.startswith('Глава'):
                s+='\n'
            if (not q.startswith('Глава')) and i!=0 and a==1:
                s+='    '
            s+=q
            s+='\n'
            if q.startswith('Глава'):
                s+='\n'
            a = 1
        else:
            q=' '.join(text[i].split())
            if (not q.startswith('Глава')) and i!=0 and a==1:
                s+='    '
            s+=q
            s+=' '
            a = 0
print(s)
```
