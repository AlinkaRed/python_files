```python
class TriangleChecker:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    def is_triangle(self):
        if not all(isinstance(side, (int, float)) for side in (self.a, self.b, self.c)):
            return "Нужно вводить только числа!"
        if self.a <= 0 or self.b <= 0 or self.c <= 0:
            return "С отрицательными числами ничего не выйдет!"
        if (self.a + self.b > self.c) and (self.a + self.c > self.b) and (self.b + self.c > self.a):
            return "Ура, можно построить треугольник!"
        else:
            return "Жаль, но из этого треугольник не сделать."
checker1 = TriangleChecker(3, 4, 5)
print(checker1.is_triangle())

checker2 = TriangleChecker(1, 2, 3)
print(checker2.is_triangle())  

checker3 = TriangleChecker(-1, 2, 3)
print(checker3.is_triangle())  

checker4 = TriangleChecker('a', 2, 3)
print(checker4.is_triangle()) 
```

```python
from datetime import datetime
class Person:
    def __init__(self, last_name, birth_date, faculty):
        self.last_name = last_name
        self.birth_date = birth_date
        self.faculty = faculty
    def age(self):
        current_year = datetime.now().year
        birth_year = self.birth_date.year
        return current_year - birth_year
    def display_info(self):
        print(f"Last Name: {self.last_name}")
        print(f"Birth Date: {self.birth_date.strftime('%d.%m.%Y')}")
        print(f"Faculty: {self.faculty}")
        print(f"Age: {self.age()} years old")
class Applicant(Person):
    def __init__(self, last_name, birth_date, faculty):
        super().__init__(last_name, birth_date, faculty)
    def display_info(self):
        super().display_info()
        print("Status: Applicant")
class Student(Applicant):
    def __init__(self, last_name, birth_date, faculty, course):
        super().__init__(last_name, birth_date, faculty)
        self.course = course
    def display_info(self):
        super().display_info()
        print(f"Course: {self.course}")
        print("Status: Student")
class Teacher(Person):
    def __init__(self, last_name, birth_date, faculty, position, experience):
        super().__init__(last_name, birth_date, faculty)
        self.position = position
        self.experience = experience
    def display_info(self):
        super().display_info()
        print(f"Position: {self.position}")
        print(f"Experience: {self.experience} years")
        print("Status: Teacher")
```

```python
person_list = [
        Applicant("Ivanov", datetime(2005, 6, 15), "Faculty of Information Technology"),
        Student("Petrov", datetime(2003, 5, 20), "Faculty of Economics", 1),
        Teacher("Sidorov", datetime(1980, 4, 10), "Faculty of Mathematics", "Associate Professor", 15),
        Student("Kuznetsov", datetime(2002, 8, 25), "Faculty of Physics", 2),
        Applicant("Smirnova", datetime(2006, 1, 30), "Faculty of Law"),
    ]
for person in person_list:
        person.display_info()
```

```python
class PairFinder:
    def __init__(self, numbers):
        self.numbers = numbers
    def find_pair(self, target_sum):
        index_map = {}
        for index, number in enumerate(self.numbers):
            complement = target_sum - number 
            if complement in index_map:
                return index_map[complement], index
            index_map[number] = index
        return None
numbers = [2, 7, 11, 15, 3, 6]
target_sum = 9

finder = PairFinder(numbers)
result = finder.find_pair(target_sum)
result
```

```python
import math
import numpy
class M:
    def __init__(self, a, b=1):
        self.a = a
        self.b = b
    def st (self):
        print(self.a**self.b)
    def sq (self):
        print(math.sqrt(self.a))
    def f (self):
        x = 1
        for i in range(1,self.a+1):
            x*=i
        print(x)
    def isNone (self):
        if self.a==None:
            print('1')
        else:
            print('0')
a = M(3,2)
a.st()
a.sq()
a.f()
a.isNone()
```

```python
import math
class Circle:
    def __init__(self, radius):
        self.radius = radius
    def area(self):
        return math.pi * (self.radius ** 2)
    def perimeter(self):
        return 2 * math.pi * self.radius
circle = Circle(6)
print(f"Площадь круга: {circle.area()}")
print(f"Периметр круга: {circle.perimeter()}")
```
