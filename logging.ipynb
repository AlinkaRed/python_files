### Вариант 2 - сокращенный путь создания логгера (пишем в файл)

```python
import logging

logger = logging.getLogger(__name__)

FORMAT = '%(asctime)s %(levelname)-8s %(message)s'
logging.basicConfig(filename='myapp.log', level=logging.INFO, format=FORMAT)
```

```python
"""
Докстринга уровня модуля
"""
import logging

logger = logging.getLogger(__name__)

logger.setLevel(logging.INFO)

handler = logging.StreamHandler()
FORMAT = '%(asctime)s %(levelname)-8s %(message)s'
formatter = logging.Formatter(fmt=FORMAT)
handler.setFormatter(formatter)
logger.addHandler(handler)

class A:
    """Докстринг для класса А"""
    count = 0
    objects = []
    names = set()
    max_name_len = 20

    def __init__(self, name: str):
        A.count += 1
        self.n = self.count
        self.name = name
        if self.is_valid():
            A.objects.append(self)
            A.names.add(name)
            self.log()

    def is_valid(self) -> bool:
        if self.name in A.names:
            logger.error(f'Имя {self.name} уже есть, пропускаем')
            return False
        if len(self.name) > self.max_name_len:
            initial_name = self.name
            self.name = initial_name[:20].strip()
            logger.warning(f'Имя "{initial_name}" укорочено до "{self.name}"')
        return True

    def __repr__(self):
        return f'A: n={self.n}, name={self.name}'

    def log(self):
        """
        Докстринга для логгера
        """
        logger.info(f'Добавлена запись {self}')

    @classmethod
    def all(cls):
        """Докстринг для метода all()"""
        for n, obj in enumerate(cls.objects):
            # pltcz vs gtxfnftv
            print(n+1, obj)

    """Какая-то прочая докстринга"""

```

```python
logger.warn('a')
```
